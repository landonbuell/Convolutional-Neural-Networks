```python
"""
Landon Buell
16 April 2020
"""

import numpy as np
import matplotlib.pyplot as plt

#from sklearn.datasets import fetch_openml

#X,y = fetch_openml('mnist_784',version=1,return_X_y=True)
```

```python
control_activations = np.random.random((20,1))

plt.figure(figsize=(2,8))
plt.imshow(control_activations,cmap=plt.cm.bone)
plt.ylabel("Original Activations",size=40,weight='bold')
plt.xticks([])
plt.yticks([])
plt.savefig("control.png")
plt.show()
```

```python
round_activations = np.round(control_activations,0)

plt.figure(figsize=(2,8))
plt.imshow(round_activations,cmap=plt.cm.bone)
plt.ylabel("Rounded Activations",size=40,weight='bold')
plt.xticks([])
plt.yticks([])
plt.savefig("rounded.png")
plt.show()
```

```python
noise_activations = control_activations + np.random.normal(loc=0,scale=1,size=control_activations.shape)

plt.figure(figsize=(2,8))
plt.imshow(noise_activations,cmap=plt.cm.bone)
plt.ylabel("Noisy Activations",size=40,weight='bold')
plt.xticks([])
plt.yticks([])
plt.savefig("noise.png")
plt.show()
```
