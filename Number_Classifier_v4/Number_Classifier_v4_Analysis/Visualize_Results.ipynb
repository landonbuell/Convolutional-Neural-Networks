```python
"""
Landon Buell
Visualize Results of Experiment
Number Clf v4
13 April 2020
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

neurons = np.array([20,40,60,80,100,120])

frame = pd.read_csv('model_analysis.csv',index_col=0)
frame.head(10)
```

```python
# Single Layer Models
single_controls = frame.iloc[0:6]
single_round_v1 = frame.iloc[6:12]
single_round_v2 = frame.iloc[12:18]
single_gauss_v1 = frame.iloc[18:24]
# Double Layer Models
double_controls = frame.iloc[24:30]
double_round_v1 = frame.iloc[30:36]
double_round_v2 = frame.iloc[36:42]
double_gauss_v1 = frame.iloc[42:48]
```

# Single Layer Models

## Plot Training Time vs. Neurons / Layer

```python
plt.figure(figsize=(20,16))
#plt.title("Training Time",size=80,weight='bold')
plt.xlabel("Neurons/Layer",size=70,weight='bold')
plt.ylabel("Training Time",size=70,weight='bold')

metric = 'train time avg'
plt.plot(neurons,single_controls[metric],
         linestyle='--',marker='o',markersize=25,label='Control')
plt.plot(neurons,single_round_v1[metric],
         linestyle='--',marker='d',markersize=25,label='Round 50%')
plt.plot(neurons,single_round_v2[metric],
         linestyle='--',marker='s',markersize=25,label='Round 100%')
plt.plot(neurons,single_gauss_v1[metric],
         linestyle='--',marker='v',markersize=25,label='Noise 50%')

plt.vlines(0,0,120,color='black')
plt.hlines(0,0,140,color='black')
plt.xticks(fontsize=40)
plt.yticks(fontsize=40)
plt.grid()
plt.legend(loc='center right',fontsize=40)
plt.tight_layout()
plt.savefig('Single_traintime.png')
plt.show()
```

# Plot Loss Function vs. Neurons / Layer

```python
plt.figure(figsize=(20,16))
#plt.title("Loss Function Value",size=80,weight='bold')
plt.xlabel("Neurons/Layer",size=70,weight='bold')
plt.ylabel("Loss Function Value",size=70,weight='bold')

metric = 'loss value avg'
plt.plot(neurons,single_controls[metric],
         linestyle='--',marker='o',markersize=25,label='Control')
plt.plot(neurons,single_round_v1[metric],
         linestyle='--',marker='d',markersize=25,label='Round 50%')
plt.plot(neurons,single_round_v2[metric],
         linestyle='--',marker='s',markersize=25,label='Round 100%')
plt.plot(neurons,single_gauss_v1[metric],
         linestyle='--',marker='v',markersize=25,label='Noise 50%')

plt.vlines(0,0,2.5,color='black')
plt.hlines(0,0,140,color='black')
plt.xticks(fontsize=40)
plt.yticks(fontsize=40)
plt.grid()
plt.legend(loc='center right',fontsize=40)
plt.tight_layout()
plt.savefig('Single_lossvalue.png')
plt.show()
```
