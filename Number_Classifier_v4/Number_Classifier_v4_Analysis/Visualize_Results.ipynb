```python
"""
Landon Buell
Visualize Results of Experiment
Number Clf v4
13 April 2020
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

neurons = np.array([20,40,60,80,100,120])

frame = pd.read_csv('model_analysis.csv',index_col=0)
frame.head(10)
```

```python
# Single Layer Models
single_controls = frame.iloc[0:6]
single_round_v1 = frame.iloc[6:12]
single_round_v2 = frame.iloc[12:18]
single_gauss_v1 = frame.iloc[18:24]
# Double Layer Models
double_controls = frame.iloc[24:30]
double_round_v1 = frame.iloc[30:36]
double_round_v2 = frame.iloc[36:42]
double_gauss_v1 = frame.iloc[42:48]
```

# Single Layer Models

## Plot Training Time vs. Neurons / Layer

```python
plt.figure(figsize=(20,8))
plt.title("Training Time vs. Neurons per Layer",size=40,weight='bold')
plt.xlabel("Neurons/Layer",size=20,weight='bold')
plt.ylabel("Time to Train [seconds]",size=20,weight='bold')

plt.plot(neurons,single_controls['train time avg'],'o-',label='Control')
plt.plot(neurons,single_round_v1['train time avg'],'o-',label='Round 50%')
plt.plot(neurons,single_round_v2['train time avg'],'o-',label='Round 100%')
plt.plot(neurons,single_gauss_v1['train time avg'],'o-',label='Noise 50%')

plt.vlines(0,0,120,color='black')
plt.hlines(0,0,140,color='black')
plt.grid()
plt.legend()
plt.savefig('Single_traintime.png')
plt.show()
```

# Plot Loss Function vs. Neurons / Layer

```python
plt.figure(figsize=(20,8))
plt.title("Loss Function Value vs. Neurons per Layer",size=40,weight='bold')
plt.xlabel("Neurons/Layer",size=20,weight='bold')
plt.ylabel("Loss Function Value",size=20,weight='bold')

plt.plot(neurons,single_controls['loss value avg'],'o-',label='Control')
plt.plot(neurons,single_round_v1['loss value avg'],'o-',label='Round 50%')
plt.plot(neurons,single_round_v2['loss value avg'],'o-',label='Round 100%')
plt.plot(neurons,single_gauss_v1['loss value avg'],'o-',label='Noise 50%')

plt.vlines(0,0,3,color='black')
plt.hlines(0,0,140,color='black')
plt.grid()
plt.legend()
plt.savefig('Single_lossvalue.png')
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.title("Iterations to Train vs. Neurons per Layer",size=40,weight='bold')
plt.xlabel("Neurons/Layer",size=20,weight='bold')
plt.ylabel("Loss Function Value",size=20,weight='bold')

plt.plot(neurons,single_controls['iterations avg'],'o-',label='Control')
plt.plot(neurons,single_round_v1['iterations avg'],'o-',label='Round 50%')
plt.plot(neurons,single_round_v2['iterations avg'],'o-',label='Round 100%')
plt.plot(neurons,single_gauss_v1['iterations avg'],'o-',label='Noise 50%')

plt.vlines(0,0,200,color='black')
plt.hlines(0,0,140,color='black')
plt.grid()
plt.legend()
plt.savefig('Single_iterations.png')
plt.show()
```

## Average Precision vs. Neurons / Layer

```python
plt.figure(figsize=(20,8))
plt.title("Average Precision Score vs. Neurons per Layer",size=40,weight='bold')
plt.xlabel("Neurons/Layer",size=20,weight='bold')
plt.ylabel("Loss Function Value",size=20,weight='bold')

plt.plot(neurons,single_controls['average precision'],'o-',label='Control')
plt.plot(neurons,single_round_v1['average precision'],'o-',label='Round 50%')
plt.plot(neurons,single_round_v2['average precision'],'o-',label='Round 100%')
plt.plot(neurons,single_gauss_v1['average precision'],'o-',label='Noise 50%')

plt.vlines(0,0,1,color='black')
plt.hlines(0,0,140,color='black')
plt.grid()
plt.legend()
plt.savefig('Single_precisions.png')
plt.show()
```
